```python

import pandas as pd

def update_dataset(file1, file2, output_file):
    # Read both CSV files into DataFrames
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)

    # List of columns to update
    columns_to_update = [
        'aFrameLength', 'frontLength', 'wheelboxSize',
        'rearLength', 'doorPosition', 'overallChassisLength', 'ImageData'
    ]

    # Iterate over each row in df2
    for index, row in df2.iterrows():
        # Find the corresponding row in df1 with the same chassisNo
        matching_row = df1[df1['chassisNo'] == row['chassisNo']]

        # If a matching row is found, update the specified columns
        if not matching_row.empty:
            for col in columns_to_update:
                df2.at[index, col] = matching_row.iloc[0][col]

    # Save the updated df2 to the output file
    df2.to_csv(output_file, index=False)
    print(f"Dataset updated and saved to {output_file}")

file1 = '/content/weighBridgeDB_Dev.dataset.csv'
file2 = '/content/weighBridgeDB_Dev.test.csv'
output_file = 'updated_dataset2.csv'
update_dataset(file1, file2, output_file)
```
